```python
import pandas as pd
import numpy as np

# Read in the data
df = pd.read_csv('Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv')

# Sample the data to speed up computation
df = df.sample(frac=0.1, random_state=10)

df.head()
```

```python
# Remove any 'neutral' ratings equal to 3
df = df[df['reviews.rating'] != 3]

# Encode 4s and 5s as 1 (rated positively)
# Encode 1s and 2s as 0 (rated poorly)
df['Positively Rated'] = np.where(df['reviews.rating'] > 3, 1, 0)
df.head(10)
```

```python
# Most ratings are positive, so the dataset is skewed
df['Positively Rated'].mean()
```

```python
from sklearn.model_selection import train_test_split

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df['reviews.text'], df['Positively Rated'], random_state=0)
```

```python
print('X_train first entry:\n\n', X_train.iloc[0])
print('\n\nX_train shape: ', X_train.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
```

```python
# Fit the CountVectorizer to the training data specifiying a minimum 
# document frequency of 3 and extracting 1-grams and 2-grams
count_vect = CountVectorizer(min_df=3, ngram_range=(1,2)).fit(X_train)

X_train_vectorized = count_vect.transform(X_train)

len(count_vect.get_feature_names())
```

```python
model = LogisticRegression()
model.fit(X_train_vectorized, y_train)

predictions = model.predict(count_vect.transform(X_test))

# As we can see, as the dataset is skewed, it is overfitting in the test data resulting in a very low AUC score
print('AUC: ', roc_auc_score(y_test, predictions))
```

```python
feature_names = np.array(count_vect.get_feature_names())

sorted_coef_index = model.coef_[0].argsort()

print('Smallest Coefs:\n{}\n'.format(feature_names[sorted_coef_index[:10]]))
print('Largest Coefs: \n{}'.format(feature_names[sorted_coef_index[:-11:-1]]))
```

```python
from sklearn import naive_bayes
```

```python
NB_model = naive_bayes.MultinomialNB()
NB_model.fit(X_train_vectorized, y_train)
predictions = model.predict(count_vect.transform(X_test))

# As we can see, as the dataset is skewed, it is overfitting in the test data resulting in a very low AUC score
print('AUC: ', roc_auc_score(y_test, predictions))
```

```python
feature_names = np.array(count_vect.get_feature_names())

sorted_coef_index = NB_model.coef_[0].argsort()

print('Smallest Coefs:\n{}\n'.format(feature_names[sorted_coef_index[:10]]))
print('Largest Coefs: \n{}'.format(feature_names[sorted_coef_index[:-11:-1]])) # As we can see this model is sensitive to stopwords! Let's see what we can do using TF IDF
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Fit the TfidfVectorizer to the training data specifiying a minimum document frequency of 5
vect = TfidfVectorizer(min_df=3).fit(X_train)
len(vect.get_feature_names())
```

```python
X_train_vectorized = vect.transform(X_train)

NB_model = naive_bayes.MultinomialNB()
NB_model.fit(X_train_vectorized, y_train)

predictions = NB_model.predict(vect.transform(X_test))

# As we can see, as the dataset is skewed, it is overfitting in the test data resulting in a very low AUC score
print('AUC: ', roc_auc_score(y_test, predictions))
```

```python
feature_names = np.array(vect.get_feature_names())

sorted_tfidf_index = X_train_vectorized.max(0).toarray()[0].argsort()

print('Smallest tfidf:\n{}\n'.format(feature_names[sorted_tfidf_index[:10]]))
print('Largest tfidf: \n{}'.format(feature_names[sorted_tfidf_index[:-11:-1]]))
```

```python
sorted_coef_index = NB_model.coef_[0].argsort()

print('Smallest Coefs:\n{}\n'.format(feature_names[sorted_coef_index[:10]]))
print('Largest Coefs: \n{}'.format(feature_names[sorted_coef_index[:-11:-1]]))
```
